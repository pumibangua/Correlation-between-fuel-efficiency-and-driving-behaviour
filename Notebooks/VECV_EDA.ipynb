```python
data['IST_DateTime'] = pd.to_datetime(data['IST_DateTime'])
data2['IST_DateTime'] = pd.to_datetime(data2['IST_DateTime'])
data3['IST_DateTime'] = pd.to_datetime(data3['IST_DateTime'])
```

```python
sorted_data=data.sort_values(by='IST_DateTime')
sorted_data2=data2.sort_values(by='IST_DateTime')
sorted_data3=data3.sort_values(by='IST_DateTime')
```

```python
sorted_data.reset_index(drop=True, inplace=True)
sorted_data2.reset_index(drop=True, inplace=True)
sorted_data3.reset_index(drop=True, inplace=True)
```

```python
EDA_columns=['HRLFC','TotalDistance','VehicleSpeed', 'EngineSpeed', 'AccPedalPosition', 'SweetSpot', 'TopGear','EngineOilPressure','EngineCoolantTemp']
EDA_data = sorted_data[EDA_columns]
EDA_data2 = sorted_data2[EDA_columns]
EDA_data3 = sorted_data3[EDA_columns]
```

After Taking Differences in the columns - total distance and HRLFC

```python
EDA_data['Distance_Travelled'] = EDA_data['TotalDistance'].diff()
EDA_data['Fuel_consumed'] = EDA_data['HRLFC'].diff()
EDA_data2['Distance_Travelled'] = EDA_data2['TotalDistance'].diff()
EDA_data2['Fuel_consumed'] = EDA_data2['HRLFC'].diff()
EDA_data3['Distance_Travelled'] = EDA_data3['TotalDistance'].diff()
EDA_data3['Fuel_consumed'] = EDA_data3['HRLFC'].diff()
```

```python
EDA_data.drop(EDA_data.index[0])
EDA_data.reset_index(drop=True, inplace=True)
EDA_data2.drop(EDA_data2.index[0])
EDA_data2.reset_index(drop=True, inplace=True)
EDA_data3.drop(EDA_data3.index[0])
EDA_data3.reset_index(drop=True, inplace=True)
```

Fuel Economy(Km/L) = Distance Travelled(Km)/Fuel consumed(L)

```python
EDA_data['Fuel_Economy']=EDA_data['Distance_Travelled']/EDA_data['Fuel_consumed']
EDA_data2['Fuel_Economy']=EDA_data2['Distance_Travelled']/EDA_data2['Fuel_consumed']
EDA_data3['Fuel_Economy']=EDA_data3['Distance_Travelled']/EDA_data3['Fuel_consumed']
```
